```python
energy_shares = data.groupby('year')[['fossil_share_energy', 'renewables_share_energy', 'nuclear_share_energy']].mean()

# Create a stacked area plot
plt.figure(figsize=(12, 6))
plt.stackplot(energy_shares.index,
              energy_shares['fossil_share_energy'],                                  #Which energy sources (fossil fuels, renewable energy, nuclear) dominate consumption in different parts of the world? How have these trends evolved over time?#
              energy_shares['renewables_share_energy'],
              energy_shares['nuclear_share_energy'],
              labels=['Fossil Fuels', 'Renewables', 'Nuclear'])

plt.title('Global Energy Consumption Mix Over Time')
plt.xlabel('Year')
plt.ylabel('Share of Total Energy (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Let's also look at regional trends for recent year (2018)
recent_regional = data[data['year'] == 2018].groupby('country')[
    ['fossil_share_energy', 'renewables_share_energy', 'nuclear_share_energy']].mean().sort_values(
    by='fossil_share_energy', ascending=False).head(10)

plt.figure(figsize=(12, 6))
recent_regional.plot(kind='bar', stacked=True)
plt.title('Energy Mix by Country (2018) - Top 10 Fossil Fuel Users')
plt.xlabel('Country')
plt.ylabel('Share of Total Energy (%)')
```

```python
df_latest = data[data['year'] == 2018]

# Calculate mean values for major energy sources
sources = ['fossil_share_energy', 'renewables_share_energy', 'nuclear_share_energy']
source_means = df_latest[sources].mean()

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(source_means, labels=['Fossil Fuels', 'Renewables', 'Nuclear'],
        autopct='%1.1f%%', startangle=90)
plt.title('Global Energy Consumption by Source (2018)')
plt.axis('equal')
plt.show()

print("\
Energy source distribution:")
for source, value in source_means.items():
    print(source.replace('_share_energy', '').capitalize() + ":", f"{value:.1f}%")

sources = ['fossil_share_energy', 'renewables_share_energy', 'nuclear_share_energy']
source_means = data_latest[sources].mean()

# Create pie chart
plt.figure(figsize=(10, 8))
plt.pie(source_means, labels=['Fossil Fuels', 'Renewables', 'Nuclear'],
        autopct='%1.1f%%', startangle=90)
plt.title('Global Energy Consumption by Source (2018)')
plt.axis('equal')
plt.show()

print("\
Energy source distribution:")
for source, value in source_means.items():
    print(source.replace('_share_energy', '').capitalize() + ":", f"{value:.1f}%")
```

```python
data_2018 = data[data['year'] == 2018].sort_values('primary_energy_consumption', ascending=False)
top_10_countries = df_2018['country'].head(10).tolist()

# Filter data for top 10 countries and selected years
years = [1990, 2000, 2010, 2018]
df_plot = data[
    (data['country'].isin(top_10_countries)) &
    (data['year'].isin(years))
]

# Create the plot
plt.figure(figsize=(15, 8))
# Use 'df_plot' instead of 'data_plot' in the sns.barplot function call
sns.barplot(data=df_plot, x='country', y='primary_energy_consumption', hue='year')

plt.xticks(rotation=45, ha='right')
plt.title('Energy Consumption by Top 10 Countries (1990-2018)')
plt.ylabel('Primary Energy Consumption (TWh)')
plt.legend(title='Year')
plt.tight_layout()
plt.show()

# Print the actual values
print("\
Top 10 countries energy consumption (TWh) in 2018:")
print(data_2018[['country', 'primary_energy_consumption']].head(10))
```
